```python
import requests

def get_crypto_price(crypto_name):
    endpoint = f"https://api.coingecko.com/api/v3/simple/price?ids={crypto_name}&vs_currencies=usd"

    try:
        response = requests.get(endpoint)
        response.raise_for_status()
        price_info = response.json()

        if crypto_name in price_info:
            return {"success": True, "data": price_info[crypto_name]}
        return {"success": False, "error": "Cryptocurrency name not recognized."}
    except requests.exceptions.RequestException as error:
        return {"success": False, "error": f"Request error: {error}"}

def show_crypto_price(crypto_name):
    """Displays the price of a cryptocurrency in a clear format."""
    print("Cryptocurrency Price Viewer")

    result = get_crypto_price(crypto_name)

    if result["success"]:
        price_details = result["data"]
        print(f"\nPrice of {crypto_name.capitalize()}:")
        print(f"1 {crypto_name.capitalize()} = {price_details['usd']} USD")
    else:
        print(f"\nError encountered: {result['error']}")

if __name__ == "__main__":
    predefined_crypto = "bitcoin"
    show_crypto_price(predefined_crypto)
```
